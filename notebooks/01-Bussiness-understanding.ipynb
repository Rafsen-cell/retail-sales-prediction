# Business Understanding

### 01. Business Problem
Retail business need accurate sales forecasts to make better operational decisions. Without reliable predictions, companies may experience overstocking, stock shortages, or inefficient inventory management, leading to increased costs and lost sales opportunities

In this project, we aim to develop a machine learning model that predicts retail sales using historical sales data, supplier information, product characteristics, and warehouse activity. This project simuulates a real-world retail sales prediction scenario using a publicly available dataset

### 02. Business Objectives
The main objectives of this project are
- Develop a regression model to predict retail sales.
- Identify the factors that  most influence retail sales.
- Build an interactive dashboard to visualize sales performance and model predictions.

### 03. Analytical Questions
This project aims to answer the following questions:
1. Which factors have the greatest impact on tetail sales?
2. Do different suppliers conribute differently to sales performance?
3. Are there seasonal patterns in retail sales across different months?
4. How are warehouse sales and retail transfers related to retail sales?
5. Which regression model provides the most accurate sales predictions?

### 04. Success Criteria
This project will be considered succesfull if it achieves the following:
- Build a reliable regression model with good predictive performance.
- Evaluate the model usiing MAE, RMSE, and R^2 Score.
- Provide meaningful business insight from the analysis.
- Develop an interactive dashboard for exploring sales trends and predictions
