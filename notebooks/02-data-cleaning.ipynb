```python
import pandas as pd
```

```python
df = pd.read_csv(r'..\data\raw\export.csv')
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')

)
df
```

```python
print("df_shape", df.shape)
df.info()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
df["retail_sales"].unique()[:10]
```

```python
df["retail_transfers"].unique()[:10]
```

```python
df["warehouse_sales"].unique()[:10]
```

```python
columns = [
    "retail_sales",
    "retail_transfers",
    "warehouse_sales"
]

for col in columns:
    df[col] = (
        df[col]
        .str.replace(",", "", regex=False)
        .str.replace(r"\((.*)\)", r"-\1", regex=True)
        .astype(float)
    )
```

```python
df.info()
```

```python
df.to_csv(r"..\data\processed\cleaned_data.csv", index=False)
```
