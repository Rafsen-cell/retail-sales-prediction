```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("..\data\processed\cleaned_data.csv")
df
```

# total_movement

```python
df["total_movement"] = df["retail_transfers"] + df["warehouse_sales"]
df
```

# drop item_description & item_code

because it's high-cardinality text, it will be difficult to encode,we can use item_type for identifier

```python
df = df.drop(columns=["item_description", "item_code"])
df
```

# encoding

```python
df["item_type"].nunique()
```

```python
df["supplier"].nunique()
```

## one-hot encoding item_type

```python
df = pd.get_dummies(df, columns=["item_type"], drop_first=True)
```

## frequency encoding suppliers

```python
freq = df["supplier"].value_counts()

df["supplier_freq"] = df["supplier"].map(freq)
```

# save new csv file

```python
data_featured = df.to_csv(r"..\data\processed\feature_engineering_data.csv")
```
